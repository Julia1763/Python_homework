```python
#x**2=t
#y=-t**2+88*t-241
def getExtremePoint(a,b,c):
    try:
        t = -(b/2/a)
        y = (4*a*c - b**2)/4/a
        return (t,y)
    except Exception as e:
        print(str(e))
        return None
getExtremePoint(-1,88,-241)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Создаём экземпляр класса figure и добавляем к Figure область Axes
fig, ax = plt.subplots()
# Добавим заголовок графика
ax.set_title('График параболы')
plt.grid()
# Название оси X:
ax.set_xlabel('x')
# Название оси Y:
ax.set_ylabel('y')
# Начало и конец изменения значения X, разбитое на 100 точек
x = np.linspace(-20, 20, 100)
# Построение прямой
y = -1*x**4+88*x**2-241
plt.ylim([-250, 1700])
# Вывод графика
ax.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Создаём экземпляр класса figure и добавляем к Figure область Axes
fig, ax = plt.subplots()
# Добавим заголовок графика
ax.set_title('Объединенный график')
plt.grid()
# Название оси X:
ax.set_xlabel('x')
# Название оси Y:
ax.set_ylabel('y')
# Начало и конец изменения значения X, разбитое на 100 точек
x = np.linspace(-20, 20, 100)
# Построение прямой
y = -1*x**4+88*x**2-241-x**3+50*x
plt.ylim([-250, 1755])
# Вывод графика
ax.plot(x, y)
plt.show()
```
